```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
reviewsDF = pd.read_csv('beer_reviews.csv')
```

```python
reviewsDF.head()
reviewsDF.info()
```

```python
ratings = pd.DataFrame(reviewsDF.groupby('beer_beerid')['review_overall'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(reviewsDF.groupby('beer_beerid')['review_overall'].count())
ratings = ratings[ratings['num of ratings'] > 100]
ratings.head()
```

```python
ax =ratings['review_overall'].hist(bins=70)
fig = ax.get_figure()
fig.savefig('hist.jpg')
```

```python
ax = sns.jointplot(x='review_overall',y='num of ratings',data=ratings,alpha=0.5)
ax.savefig("output.jpg")
```

```python
filteredDF = pd.merge(ratings['num of ratings'], reviewsDF, on='beer_beerid')
filteredDF.head()
```

```python
beermat = filteredDF.pivot_table(index='review_profilename',columns='beer_beerid',values='review_overall')
beermat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
beer_rating = beermat[2093]
beer_rating.head()
```

```python
similar_to_beer = beermat.corrwith(beer_rating)
```

```python
corr_beer = pd.DataFrame(similar_to_beer,columns=['Correlation'])
corr_beer.dropna(inplace=True)
corr_beer.head()
```

```python
corr_beer.sort_values('Correlation',ascending=False).head(10)
```
